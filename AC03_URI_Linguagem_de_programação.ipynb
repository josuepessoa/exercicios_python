```python
#@title 
#beecrowd | 1269

# Função utilizada para impressao de resultados formatado
def exibir(dados_exibicao): 
  # Pega array de dados em formato numerico e coloca em ordem crescente
  dados_exibicao = sorted(dados_exibicao) 

  # Pega o array e transforma dados numericos em string
  dados_exibicao = [str(dado) for dado in dados_exibicao]

  # Pega o array e devolve em uma string separada por espaço
  mensagem = ' '.join(dados_exibicao)
  print(mensagem) 

# Inicia execução principal recebendo primeiro input
dados = input()

# Transforma input string em array ja no formato numerico
# O formato está numerico pois é mais facil para ordenar
dados= [int(n) for n in dados.split()]

# Começa chacagens da execução principal
while True:

  try:  
    # Pega input para verificar qual ação será executada
    acao = input().split()  

    # Se a ação for de Encerrar, 
    # chama função para exibir os dados e depois encerra
    # completamente a execução 
    if acao[0] == "encerrar":
      exibir(dados)  
      break

    # Se a ação for Exibir, chama a função que exibe os dados 
    if acao[0] == "exibir":
      exibir(dados)  

    # Se a ação for Adicionar, adiciona o item ao array de dados 
    if acao[0] == "adicionar":
      dados.append(int(acao[1]))

    # Se a ação for Remover,
    if acao[0] == "remover":

      # Variavel de controle, pois quando não encontrar o dado a ser
      # removido no array de dados, é preciso exibir mensagem
      achou_dado = False
      acao[1] = int(acao[1])
      # Percorre todo o array de dados, passando por cada dado      
      for dado in dados:
        # Caso encontre o dado, marca a variavel achou dado como True
        if dado == acao[1]:
          achou_dado = True
      
      # No fim da execução do laço, verifica se encontrou o dado a ser
      # apagado, caso encontre apaga o dado do array de dados.
      if achou_dado:
        dados.remove(acao[1])
      #caso não encontre o dado, exibe a mensagem.
      else:
        print(f"código {acao[1]} não encontrado")

  # Caso de erro na tentativa principal, força parada da execução      
  except EOFError:
    break
```

```python
#beecrowd | 1270

#variaveis de inicio
tabuada_inicio = int(input())
tabuada_fim = int(input())

# Se a tabuada final for maior ou igual a tabuada inicial, faz o calculo.
if tabuada_fim >= tabuada_inicio:

  # Para cada tabuada encontrada entre a tabuada_inicio e a tabuada_final +1.
  # Feito a soma de 1 na tabuada final pois o FOR RANGE executa até menor que
  # o final da função range e não menor e igual.
  for tabuada in range(tabuada_inicio, tabuada_fim + 1):

    # Para cada multiplicador encontrado entre 1 e menor que 10.
    # Range contem 11 pois sua execução é menor e não menor e igual
    for multiplicador in range(1,11):
      print(f"{tabuada} x {multiplicador} = {tabuada * multiplicador}")
      
    # No fim de cada tabuada, imprime o separador
    print('-'*10)
# Cai no SE NÃO, caso a tabuada final seja menor que a tabuada inicial.
else:
  print("Nenhuma tabuada no intervalo!")
```

```python
# beecrowd | 1282

from math import floor

# Coleta numero de dados de canais a serem processador
numero_dados = int(input())
contador = 1
informacao_canais= []

# Proteção para pegar apenas dados entre 1 e 200
if 1 <= numero_dados <= 200:
  
  # Coleta dados do canal para cada linha digitada.
  while contador <= numero_dados:
    dados = input().split(";")
    dados[1] = int(dados[1])
    dados[2] = round(float(dados[2]),2)
    informacao_canais.append(dados)
    contador +=1
  
  # Coleta valores premium e normal
  premium = float(input())
  normal = float(input())
  contador = 0 

  # Imprime cabeçlho do exercicio
  print("-----")
  print("BÔNUS")
  print("-----")

  # Executa verificação para todos os canais, do array informacao_canais
  while contador < len(informacao_canais):

    # Calcula qual foi o numero de cpm
    cpm = floor(informacao_canais[contador][1]/1000)

    # Caso tenha conteudo premium, usa taxa premium, se nao usa taxa normal
    monetizacao = cpm * premium if informacao_canais[contador][3] == "sim" else cpm * normal

    # Calcula valor a receber
    valor_receber = monetizacao + informacao_canais[contador][2]

    # Imprime informação formatada   
    print(f"{informacao_canais[contador][0]}: R$ {valor_receber:.2f}")

    # Atualiza contador
    contador += 1
  
```

```python
#beecrowd | 1283

#Variaveis iniciais
tempo_corrida = 1 
tempos_corridas = []

# Enquanto o tempo da corrida for positivo, continuaremos a capturar o tempo da corrida
while tempo_corrida > 0 :

  # Captura o tempo da corrida
  tempo_corrida = int(input())

  # Caso o tempo da corrida for maior que zero, inclui no array de tempos de corridas
  if tempo_corrida > 0:
    tempos_corridas.append(tempo_corrida)

# Calcula a media com a somatoria das corridas dividido pela quantidade de corridas armazenadas
media = (sum(tempos_corridas) / len(tempos_corridas))
print(f"MEDIA: {media:.2f}")

# Percorre todas as corridas armazenadas no array tempos_corridas
for tempo_corrida in tempos_corridas:

  # Imprime o tempo da corrida caso seja menor que a média calculada.
  if tempo_corrida < media :
    print(tempo_corrida)
```

```python
#beecrowd | 1284

quantidade_alunos = int(input())
notas = []

if 1 <= quantidade_alunos <= 999:
    # Armazenando notas iniciais dos alunos                         
    for aluno in range(quantidade_alunos):
        nota_inicial = float(input())
        notas.append([nota_inicial])
    
    # Armazenando segundas nostas dos alunos
    for aluno in range(quantidade_alunos):
        segunda_nota = float(input())
        notas[aluno].append(segunda_nota)
    
    # Verificando se houve notas alteradas de acordo com as regras
    alteradas = 0
    for alunos in range(quantidade_alunos):
        # Se a nota ja for 10 não muda nada
        if notas[alunos][0] == 10:
            notas[alunos].append(notas[alunos][0])
            #marcador de nota não alterada será inserido no vetor
            notas[alunos].append('-')
                        
        # Caso a segunda nota seja 10, acrescenta 2 pontos na nota limitando a 10 e marcando como nota alterada "*"
        elif notas[alunos][1] == 10:
            nota_nota = notas[alunos][0] + 2
            if nota_nota > 10:
                nota_nota = 10                
            notas[alunos].append(nota_nota)
            notas[alunos].append('*')
            alteradas +=1

        # Não altera nada    
        else:
            notas[alunos].append(notas[alunos][0])
            notas[alunos].append('-')

    #Mostra quantas notas foram alteradas
    print(f'NOTAS ALTERADAS: {alteradas}')

    # Varre todas as notas mostrando o resultado
    for aluno in range(quantidade_alunos):
        print(f'{notas[aluno][3]}({aluno+1:03}) original: {notas[aluno][0]:05.2f} | final: {notas[aluno][2]:05.2f}')
```
